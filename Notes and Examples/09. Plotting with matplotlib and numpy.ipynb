## Basic Plotting

Now that we have imported our library, let's get right to it and learn how to construct some basic plots.  Today, we will see how to plot *scatter plots*, *histograms* and *lines*.  All three have very similar syntax, it's just remembering how to call the specfic plot that you'd like.

### Lines

To construct a plot that connects each point with a line, we will use the `plot(x,y)` function in the library.

```python
# Construct two lists - x and y - that will give the points on the graph
x = [1,2,3,4,5]
y = [3,4,8,9,10]

# Construct the plot
plt.plot(x,y)

# "show" the plot as the cell's output
plt.show()
```

This is a pretty lame graph, and needs some additional information to make it complete.  We need to be sure to add in an $x$-label, a $y$-label and a title - especially if this is a graph of important information.

```python
# Construct two lists - x and y - that will give the points on the graph
x = [1,2,3,4,5]
y = [3,4,8,9,10]

# Construct the plot
plt.plot(x,y)

# Adding in x-, y-labels and a title - label need to come AFTER you contruct plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

plt.show()
```

The `plot()` function allows for many other arguement, not just `x` and `y`, as we've done above.  A third arguement can be placed which is often labelled as the **format**.  The format describes the style of line and the colour you wish to use.  You can also specify in the arguement the **line width** using the syntax `linewidth = value`, where `value` is how thick you'd like the line.  To get all the intricate details of `plot()`, check out the webpage: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# Construct two lists - x and y - that will give the points on the graph
x = [1,2,3,4,5]
y = [3,4,8,9,10]

# Construct the plot using green (g) circles (o) connected by a solid line (-)
plt.plot(x,y,'go-', linewidth = 3) 
#for the third argument, in quotations we specifify: [COLOUR - MARKER - LINE TYPE]
#for the fourht arguement we specify the linewidth using: [linewidth = value]

# Adding in x-, y-labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

plt.show()
```

### Scatter plots

In `matplotlib.pyplot`, to construct a scatter plot we use the function `scatter(x,y)`.  Note, just like `plot()`, `scatter()` has additional "hidden" agruements that we can adjust.

```python
# Data for our scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 9]

# Construct our scatter plot
plt.scatter(x,y,color = 'black', marker = '^', s = 50) 
#'color' changes the... colour
#'marker' changes the plotting symbol
#'s' is for size (default is 50)

#blahblah labels
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Wewoo')

# show the scatterplot
plt.show()
```

### Multiple plots

We can also place multiple plots onto a single visualization, if we so choose.  For instance, if we have two data sets that we wish to visualize together, we simply tell Python to plot both.

```python
# Data for our scatter plot
x = [1, 2, 3, 4, 5]
y1 = [4, 5, 9, 10, -2]
y2 = [9, -3, 8, 1, 4]

# use the scatter plot command twice
plt.scatter(x,y1,color = 'black', marker = 's')
plt.scatter(x,y2, color = 'orange', marker = '*')

# labels and a legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Wewoo')
plt.legend(['Data set 1','Data set 2'], loc = 'lower center')

#saving the figure - must come BEFORE show figure
plt.savefig("samplemultiplot.png")

# show the figure
plt.show()
```

### Histograms

Histograms are used to visualize the frequency of a variety of data points - at least in Tyler's mind... don't tell a stats prof this.  They use slightly different syntax than regular plots, but the function we use is `hist()`.

```python
# Given data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7]

# Create a histogram 
plt.hist(data, bins = 10, edgecolor = 'y')
#'bins' is for number of bars
#'edgecolor' change the outline of bars

# Create labels
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")

# Show histogram
plt.show()
```

## Numpy and Matplotlib
> Using numpy arrays to assist us in plotting complicated functions

---

You can use `numpy` arrays to generate data for `matplotlib` plots easily.  We commonly generate a lot of $x$ values in an array, and then compute the associated $y$ values using some mathematical equation.  To generate **a lot** of $x$ values extremely quickly, we introduce a new function in `numpy` called `linspace(a,b,n)`.  This will generate $n$ equally spaced values in the interval from $[a,b]$ in an array.

```python
# generate x values between 0 and 10 and y values that are equal to x^2 using n = 5
x = np.linspace(0,10,5) #increasing the n-value will smooth out the line
y = x ** 2

# plot these values
plt.plot(x,y)

# show the plot
plt.show()
```
