# Introduction to Pandas


Pandas is a powerful data analysis and manipulation library for Python. It provides data structures like Series and DataFrame, which are flexible and easy to use for handling large datasets. 

Key Features:
- Easy handling of missing data
- Flexible reshaping and pivoting of datasets
- Label-based slicing, indexing, and subsetting of large datasets
- Data alignment and integration of data from different sources


# Importing and Plotting with Pandas
Pandas integrates well with Matplotlib, allowing for easy plotting of data directly from DataFrames and Series.

```python
# Importing Libraries:
import pandas as pd
import matplotlib.pyplot as plt
```

# Pandas Series (LIST)
- one-dimensional array-like object that can hold data of any type (integers, strings, floating points, etc.).
- It is similar to a column in a spreadsheet or a database table.


```python
# Creating a Series from a list
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(data)
print("")
print(series)
```

# Pandas Series (Dictionary)
A Series can also be created from a dictionary, where the keys become the index, and the values are the data.

```python
# Creating a Series from a dictionary
data = {'a': 10, 'b': 20, 'c': 30}
series = pd.Series(data)
print(series)
```

# Series Properties
Pandas Series have several useful properties:
- **`index`**: Returns the index (row labels) of the Series.
    - The index can be customized, or it will default to a 'RangeIndex' if none is provided.
- **`values`**: Returns the data as a NumPy array.
- **`dtype`**: Returns the data type of the values.

```python
series = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(series.index)
print(series.values)
print(series.dtype)
```

# Detecting Missing Data
Pandas provides functions to detect and handle missing data, such as `isna()`, `notna()`, and `fillna()`.

```python
# Series with missing data
data = [1, None, 3, None, 5]
series = pd.Series(data)

# Detect missing values
print(series.isna())

# Fill missing values
filled_series = series.fillna(0)
print(filled_series)
```

# Pandas DataFrame
A DataFrame is a two-dimensional labeled data structure with columns of potentially different types. It is similar to a table in SQL or a spreadsheet data representation.

{'A': ['A0', 'A1'], 'B': ['B0', 'B1']}: This is a dictionary where:
Keys ('A' and 'B'): These represent the column names of the DataFrame.
Values (['A0', 'A1'] and ['B0', 'B1']): These are lists that contain the data for each column.

```python
# Creating a DataFrame from a dictionary of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}
df = pd.DataFrame(data)
print(df)
```

## Adding Data to a DataFrame
Data can be added to a DataFrame using `loc`, `iloc`, or `assign()` methods.

```python
# Adding a new column to a DataFrame
df['Department'] = ['HR', 'Finance', 'Engineering']
print(df)

# Adding a new row
df.loc[3] = ['David', 40, 80000, 'Marketing']
print(df)
```

## Selecting Data
- **`loc[]`**: Selects data based on labels (index and column names).
- **`iloc[]`**: Selects data based on integer positions.

```python
df = pd.DataFrame({
    'Company': ['Daimler', 'E.ON', 'Siemens', 'BASF', 'BMW'],
    'Price': [69.2, 8.11, 110.92, 87.28, 87.81],
    'Volume': [4456290, 3667975, 3669487, 1778058, 1824582]
}, index=['a', 'b', 'c', 'd', 'e'])
```

```python
# Selecting rows (c, d, and e) and columns (Volume and Price) with loc
print(df.loc[['c', 'd', 'e'], ['Volume', 'Price']])

# Selecting rows from 'b' to 'd' and all columns
print(df.loc['b':'d', :])

# Selecting rows where Price is greater than 50
print(df.loc[df['Price'] > 50])

# Selecting.. 1st and 3rd rows [0,2] and columns 1 and 2 [0,1]
print(df.iloc[[0, 2], [0, 1]])
```

## Sorting Data
- with `sort_values()`

```python
# Sorting by 'Price' in ascending order
sorted_df = df.sort_values(by='Price')
print(sorted_df)

# Sorting by 'Price' in descending order
sorted_df_desc = df.sort_values(by='Price', ascending=False)
print(sorted_df_desc)
```

```python
df = pd.DataFrame({
    'A': [3, 1, 2],
    'B': [9, 7, 8]
}, index=['c', 'a', 'b'])

sorted_df = df.sort_index()  # Sorts rows by index
print(sorted_df)
```

## Grouping Data 
- with `groupby()`
- Groups rows based on one or more columns, then applies aggregation functions like `sum()`, `mean()`, etc.

```python
data = {
    'Company': ['Daimler', 'Daimler', 'BMW', 'BMW', 'Siemens'],
    'Year': [2020, 2021, 2020, 2021, 2020],
    'Sales': [100, 150, 200, 250, 300]
}
df = pd.DataFrame(data)
```

```python
grouped = df.groupby(['Company', 'Year']).sum()  # Groups by 'Company' and 'Year' and sums values
print(grouped)
```

## Merging DataFrames 
- with `merge()`
- Combines two DataFrames using a common column.

```python
df1 = pd.DataFrame({
    'Company': ['Daimler', 'BMW', 'Siemens'],
    'Revenue': [100, 200, 300]
})

df2 = pd.DataFrame({
    'Company': ['Daimler', 'BMW', 'BASF'],
    'Profit': [10, 20, 30]
})

merged_df = pd.merge(df1, df2, on='Company', how='inner')  # Merges DataFrames on 'Company'
print(merged_df)
```

## Concatenating DataFrames 
- with `concat()`
- Joins multiple DataFrames along rows or columns.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})

concatenated = pd.concat([df1, df2], axis=0)  # Concatenates DataFrames vertically
print(concatenated)
```

## Pivoting Data 
- with `pivot_table()`
- **Pivot Table**: Summarizes data by reshaping and aggregating.

```python
df = pd.DataFrame({
    'Company': ['Daimler', 'BMW', 'BMW', 'Siemens', 'Siemens'],
    'Year': [2020, 2020, 2021, 2020, 2021],
    'Sales': [100, 200, 150, 300, 400]
})

pivot = df.pivot_table(values='Sales', index='Company', columns='Year', aggfunc='sum')  # Creates a pivot table
print(pivot)
```

## Applying Functions to DataFrames 
- with `apply()`
- **Applying Functions**: Applies a function across rows or columns.

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df['Sum'] = df.apply(lambda row: row['A'] + row['B'], axis=1)  # Adds A and B row-wise
print(df)
```

## Handling Missing Data 
- with `dropna()` and `fillna()`
- **Removing/Handling Missing Data**: Drops rows/columns with missing values or fills them.

```python
df = pd.DataFrame({
    'A': [1, None, 3],
    'B': [4, 5, None]
})

df_dropped = df.dropna()  # Removes rows with missing values
df_filled = df.fillna(0)  # Fills missing values with 0
print(df_dropped)
print(df_filled)
```

## Renaming Columns 
- with `rename()`
- Changes the names of columns using a dictionary.

```python
df = pd.DataFrame({
    'old_name': [1, 2, 3],
    'another_old_name': [4, 5, 6]
})

df_renamed = df.rename(columns={'old_name': 'new_name'})  # Renames column 'old_name' to 'new_name'
print(df_renamed)
```

## Dropping Duplicates 
- with `drop_duplicates()`
- Removes duplicate rows based on specified columns.

```python
df = pd.DataFrame({
    'A': [1, 2, 2, 3],
    'B': [4, 5, 5, 6]
})

df_unique = df.drop_duplicates()  # Drops rows with duplicate values
print(df_unique)
```

## Resetting Index 
- with `reset_index()`
- Resets the index of the DataFrame, turning the current index into a column.

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}, index=['x', 'y', 'z'])

reset_df = df.reset_index()  # Resets the index and turns it into a column
print(reset_df)
```

## Calculating Bollinger Bands
Steps:
1. Calculate Moving Average
2. Calculate Standard Deviation
3. Calculate Upper and Lower Bollinger Bands

```python
import yfinance as yf
import pandas as pd

# Fetch stock data
stock = yf.download("F", start="2020-01-01", end="2020-08-31")

# Calculate the 20-day moving average and standard deviation
stock['MA20'] = stock['Close'].rolling(window=20).mean() #20-day MA
stock['STD20'] = stock['Close'].rolling(window=20).std() #20-day SD

# Calculate upper and lower bands
stock['Upper_Band'] = stock['MA20'] + (2 * stock['STD20'])
stock['Lower_Band'] = stock['MA20'] - (2 * stock['STD20'])

stock
```

## Plotting Price Chart w/ Bollinger Bands


```python
import matplotlib.pyplot as plt

# Plot the adjusted close price and Bollinger Bands
plt.figure(figsize=(14, 7))
plt.plot(stock['Adj Close'], label='Adjusted Close Price', color='blue')
plt.plot(stock['MA20'], label='20-Day MA', color='green')
plt.plot(stock['Upper_Band'], label='Upper Band', linestyle='--', color='red')
plt.plot(stock['Lower_Band'], label='Lower Band', linestyle='--', color='red')
plt.fill_between(stock.index, stock['Upper_Band'], stock['Lower_Band'], color='red', alpha=0.1)
plt.title('Ford Stock Price with Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

## Comparing Stock Trends 
Steps:
1. Standardize the data = `(price - mean price) / std dev`
2. Plot the standardized prices of two stocks

```python
msft = yf.download("MSFT", start="2020-01-01", end="2020-08-31")
aapl = yf.download("AAPL", start="2020-01-01", end="2020-08-31")

# Standardize the adjusted close price
msft['Standardized'] = (msft['Adj Close'] - msft['Adj Close'].mean()) / msft['Adj Close'].std()
aapl['Standardized'] = (aapl['Adj Close'] - aapl['Adj Close'].mean()) / aapl['Adj Close'].std()

# Plot the data
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_ylabel("Standardized Price")
msft['Standardized'].plot(ax=ax, label="Microsoft")
aapl["Standardized"].plot(ax=ax, label="Apple")
ax.legend(loc="best")

# Save the plot
#fig.savefig("out/compare.pdf")
```

### Percentage-Based Normalization

```python
# Normalize the data to start at 100 for both stocks
msft_normalized = msft / msft.iloc[0] * 100
aapl_normalized = aapl / aapl.iloc[0] * 100

# Plot the normalized data
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_ylabel("Percentage")
msft_normalized['Adj Close'].plot(ax=ax, label="Microsoft")
aapl_normalized['Adj Close'].plot(ax=ax, label="Apple")
ax.legend(loc="best")

# Save the plot
#fig.savefig("out/comparenew.pdf")
```

# Calculating Stock Returns
`pct_change`

```python
# Calculate daily returns
stock['Daily_Return'] = stock['Adj Close'].pct_change()
```

## Calculating and Plotting Cumulative Returns
Steps:
1. Calculate cumulative returns by compounding daily returns.
2. Plot the cumulative returns.


```python
# Calculate cumulative returns
stock['Cumulative_Return'] = (1 + stock['Daily_Return']).cumprod()
```

```python
# Plot cumulative returns
plt.figure(figsize=(14, 7))
plt.plot(stock['Cumulative_Return'], label='Cumulative Return', color='purple')
plt.title('Cumulative Returns of Ford Stock')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
```

# Volatility
Steps:
1. Use the rolling standard deviation of daily returns to measure volatility.


```python
# Calculate 20-day rolling volatility
stock['Volatility'] = stock['Daily_Return'].rolling(window=20).std() * (252**0.5)  # Annualized volatility
```

```python
# Plot volatility
plt.figure(figsize=(14, 7))
plt.plot(stock['Volatility'], label='Volatility', color='red')
plt.title('20-Day Rolling Volatility of Ford Stock')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```

# OLS Regression
- Using; `statsmodels`

```python
import statsmodels.api as sm
```

```python
n = 100
beta_0 = 5
beta_1 = 3
sigma = 2

#Simulate data
x = np.random.uniform(0,10, size=n)
epsilon = np.random.normal(0, sigma, size=n)
y = beta_0 + beta_1 * x + epsilon

#OLS estimation
X = sm.add_constant(x) #Add intercept
model = sm.OLS(y, x)
results = model.fit()

print(results.summary())
```
